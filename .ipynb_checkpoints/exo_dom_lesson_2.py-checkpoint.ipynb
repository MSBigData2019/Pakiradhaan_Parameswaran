Pour les sociétés suivantes : Aribus, LVMH et Danone : 

* les ventes au quartier à fin décembre 2018 ;
* le prix de l'action et son % de changement au moment du crawling ;
* le % Shares Owned des investisseurs institutionels ;
* le dividende yield de la company, le secteur et de l'industrie.

```python
import urllib.request as urllib
from bs4 import BeautifulSoup
```

```python
quote_page = "https://www.reuters.com/finance/stocks/financial-highlights/LVMH.PA"
page = urllib.urlopen(quote_page)
```

```python
soup = BeautifulSoup(page, 'html.parser')
```

```python
print(soup)
```

```python
float(soup.find("td", string='Quarter Ending\xa0Dec-18').parent.find_all("td")[2].text.replace(',',''))
```

```python
float(soup.find("div", class_="sectionQuoteDetail").find_all("span")[1].text.strip())
```

```python
soup.find("span", class_="valueContentPercent").find("span", class_="pos").text.strip().replace('(','').replace(')','')
```

```python
soup.find("strong", string="% Shares Owned:").findNext().text
```

```python
dividende = soup.find("td", string="Dividend Yield").parent.find_all("td", class_="data")
for div in dividende :
    print(float(div.text))
```
