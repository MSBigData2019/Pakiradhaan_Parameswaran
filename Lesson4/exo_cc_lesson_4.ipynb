## Infos sur les paracétamols

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import numpy as np
import re
```

```python
url = "https://www.open-medicaments.fr/api/v1/medicaments?query=paracetamol"
```

```python
req = requests.get(url)
df = pd.read_json(req.content)
df = df.denomination.str.extract(r'([\D]*)(\d+)(.*),([\w\s]*)')
df[2] = df[2].apply(lambda x : x.strip())
df['coef'] = df[2].apply(lambda x: 1000 if x=="g" else 1)
df['dose'] = df['coef'] * df[1].apply(lambda x : int(x.strip()))
```
