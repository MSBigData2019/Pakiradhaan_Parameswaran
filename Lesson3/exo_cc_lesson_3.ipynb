## Matrice de distances entre les 10 plus grandes villes de France

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import numpy as np
import json
```

```python
key = open("key.txt", "r").read()
gapi_url = "https://maps.googleapis.com/maps/api/distancematrix/json?units=imperial&origins={}&destinations={},NY&key={}"
city_url = "https://fr.wikipedia.org/wiki/Liste_des_communes_de_France_les_plus_peupl%C3%A9es"
```

```python
def get_soup_from_link(link) :
    hdr = {'User-Agent': 'Mozilla/5.0'}
    req = requests.get(link, headers=hdr)
    soup = BeautifulSoup(req.content, 'html.parser')
    return soup
```

```python
table = get_soup_from_link(city_url).find('table').findAll('tr')
list_cities = list(map(lambda x : table[x].find('a').text, range(1,11)))
```

```python
response = requests.get(gapi_url.format('|'.join(list_cities), '|'.join(list_cities), key))
json_response = json.loads(response.content)
```

```python
distances = list(map(lambda x: list(map(lambda y: y['distance']['value'] if y['status']=='OK' else 'KO' ,x['elements'])), json_response['rows']))
```

```python
mat_dist = pd.DataFrame(columns=list_cities, index=list_cities, data=distances)
mat_dist
```
